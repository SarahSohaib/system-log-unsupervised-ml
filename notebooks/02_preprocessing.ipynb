# System Log and Resource Usage Pattern Analysis using Unsupervised Machine Learning
## Notebook 02: Data Preprocessing

This notebook details the crucial preprocessing steps required for analyzing system logs and resource usage patterns using unsupervised machine learning. Effective preprocessing ensures data quality, reduces noise, and transforms raw data into a suitable format for algorithms like PCA and K-Means clustering, thereby enhancing the accuracy and interpretability of the subsequent analysis.

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
import os
os.chdir('/content/drive/MyDrive/system-log-unsupervised-ml')
print(os.getcwd())
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("data/raw/system_metrics.csv")
df.head()
```

### Preprocessing Strategy

This section outlines the strategic approach to preprocessing the system metrics dataset for unsupervised learning. The dataset comprises various time-series system performance indicators, including CPU utilization, memory consumption, disk I/O operations, network activity, and overall system load. Given that the dataset is synthetically generated, it inherently exhibits an absence of missing values or data inconsistencies, simplifying the initial data cleaning phase.

For the purpose of unsupervised clustering and dimensionality reduction, only numerical features directly representative of system behavior are retained. The 'timestamp' column is explicitly excluded from the feature set, as its temporal nature does not contribute directly to the pattern identification objectives of clustering or principal component analysis.

Furthermore, feature scaling is an indispensable step prior to applying distance-based algorithms such as PCA and K-Means. Unscaled features, particularly those with disparate magnitudes or units, can lead to biased results, where features with larger numerical ranges disproportionately influence the distance calculations. Therefore, standardization will be applied to ensure all features contribute equally to the analytical process.

```python
features = df.drop(columns=["timestamp"])
features.head()
```

```python
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)
scaled_df = pd.DataFrame(scaled_features, columns=features.columns)
scaled_df.head()
```

```python
output_dir = "data/processed"
os.makedirs(output_dir, exist_ok=True)
scaled_df.to_csv(os.path.join(output_dir, "system_metrics_scaled.csv"), index=False)
print(f"Processed data saved to: {os.path.join(output_dir, 'system_metrics_scaled.csv')}")
```

### Key Preprocessing Observations

*   **Timestamp Column Removed:** The 'timestamp' column, being non-contributory to clustering, has been successfully excluded from the feature set.
*   **Numerical Data Retention:** The dataset now exclusively contains numerical features, which are appropriate for the chosen unsupervised learning algorithms.
*   **Feature Scaling Applied:** All features have undergone standardization using `StandardScaler`, ensuring uniform contribution to distance-based models.
*   **Dataset Prepared for Analysis:** The preprocessed dataset is now in a clean, scaled format, making it suitable for subsequent dimensionality reduction (e.g., PCA) and clustering (e.g., K-Means) tasks.
*   **Data Persistence:** The scaled dataset has been saved, ensuring reproducibility and availability for subsequent analytical notebooks.
